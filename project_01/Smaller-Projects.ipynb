```python
#reproduce the polynomial x**2 function plot
polynomialx2=pd.DataFrame({'x':range(-10,11),'y':[i ** 2 for i in range (-10,11)]}) # calculate second powers via list comprehension

polynomialx2.plot('x','y',color = 'green')
plt.axis([-11.0, 11.0, -10.0, 110.0])
plt.xlabel('x')
plt.title('Polynomial y=x²')
plt.legend(loc='lower left')

plt.savefig('polynomialx**2.png', dpi=200)

plt.show()
```

```python
# create fertility bar plots
dffertile = pd.read_csv('data/gapminder_total_fertility.csv', index_col=0)
dffertilesample = dffertile[['1810','1990','2010']].loc[['Germany','Turkey','Zimbabwe']]
dffertilesample.plot.bar(figsize=(7,5))
plt.xticks(rotation=0)                    # fix x label rotation
plt.savefig('1810-1990-2010-fertilities.png')
dffertilesample.plot.bar(y='1990', width=0.7, color='orange')
plt.xticks(rotation=0)
plt.savefig('1990-fertilities.png')
dffertilesample.transpose().plot.bar(figsize=(7,5))
plt.xticks(rotation=0)
plt.savefig('1810-1990-2010-fertilities-transposed.png')
```
